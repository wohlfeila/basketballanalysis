```python
# Import packages
import numpy as np
import random
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import warnings
import re
```

```python
# Ignore warnings
warnings.filterwarnings('ignore')
```

```python
nltk.download('punkt', quiet = True)
nltk.download('wordnet', quiet = True)
```

```python
f = open('timberwolves_history.txt', 'r')
data = f.read()
```

```python
text = re.sub("\[\d+\]", "", data)
```

```python
sent_tokens = nltk.sent_tokenize(text)

print(sent_tokens)
```

```python
remove_punct_dict = dict((ord(punct),None) for punct in string.punctuation)

print(string.punctuation)

print(remove_punct_dict)
```

```python
# create a function to return a list of lemmtized lower case words after removeing punctuations
def tokenization(text):
    return nltk.word_tokenize(text.lower().translate(remove_punct_dict))

print(tokenization(text))
```

```python
# keyword matching

# greeting inputs
GREETING_INPUTS = ["hi", "hello", "sup", "hey", "yo"]

# greeting responses
GREETING_RESPONSES = ["hi", "howdy", "hello", "hey", "hi there"]

# function to return random greeting response
def greeting(sentence):
    # if users input is a greeting, then return a randomly chosen greeting reponse
    for word in sentence.split():
        if word.lower() in GREETING_INPUTS:
            return random.choice(GREETING_RESPONSES)
```

```python
def response(user_response):
    # user input
    #user_response = "who is the coach of the timberwolves"

    # make response lower case
    user_response = user_response.lower()

    # chatbot response
    robo_response = ''

    # append users response to sentences
    sent_tokens.append(user_response)

    # create tfidf vecotrizer
    TfidfVec = TfidfVectorizer(tokenizer = tokenization, stop_words = 'english')

    # fit tfidfvec
    tfidf = TfidfVec.fit_transform(sent_tokens)

    # get similarity
    vals = cosine_similarity(tfidf[-1], tfidf)

    # get index of most similarity
    idx = vals.argsort()[0][-2]

    flat = vals.flatten()

    flat.sort()

    score = flat[-2]

    if(score == 0):
        robo_response = robo_response+"Please ask another question."
    else:
        robo_response = robo_response+sent_tokens[idx]

    sent_tokens.remove(user_response)

    return(robo_response)
```

```python
flag = True
print("TimberBot: I am TimberBot. I will answer questions you have about the Timberwolves")
while(flag == True):
    user_response = input()
    user_response = user_response.lower()
    if(user_response != 'bye'):
        if(user_response == 'thanks' or user_response == 'thank you'):
            flag=False
            print("TimberBot: You're welcome!")
        else:
            if(greeting(user_response) != None):
                print("TimberBot: "+greeting(user_response))
            else:
                print("TimberBot: "+response(user_response))
    else:
        flag = False
        print("TimberBot: Chat with you later!")
```
